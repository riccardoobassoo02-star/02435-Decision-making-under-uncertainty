```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_and_show_statistics(file_path):
    data = np.loadtxt(file_path, delimiter=",", skiprows=1)
    
    means = data[:, 0]
    stds  = data[:, 1]
    max_time = data[:, 2]
    avg_time = data[:, 3]

    print("Average objective function:", round(np.mean(means), 2))
    print("Std of objective function:", round(np.mean(stds), 2))
    print(f"Max policy time: {round(np.max(max_time), 2)}s")
    print(f"Average policy time: {round(np.mean(avg_time), 2)}s")
    
    return means, stds, max_time, avg_time
```

L=5 & B=3 when there are more than 1 active overrides

```python
stats = load_and_show_statistics("Policy_log_files/SP_policy_30_v2_log_100_exp_4_reps_1.csv")
```
